```python
import requests

url = "https://quotes.toscrape.com"

response = requests.get(url)
print(response.text)

```

```python
from bs4 import BeautifulSoup

html = response.text
soup = BeautifulSoup(html,"html.parser")
print(soup)
```

```python
quote = soup.find("span", class_="text")
print(quote.text)
```

```python
quotes = soup.find_all("span", class_="text")
for q in quotes:
  print(q.text)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://quotes.toscrape.com"

html = requests.get(url).text

soup = BeautifulSoup(html, "html.parser")

quotess = soup.find_all("div", class_="quote")

for quote in quotess:

    text = quote.find("span", class_="text").text

    author = quote.find("small", class_="author").text

    print(text)

    print(author)

    print("-"*50)
```

```python
import pandas as pd

data = []

quotess = soup.find_all("div", class_="quote")
for w in quotess:
  text = w.find("span", class_="text").text
  author = w.find("small").text
  data.append([text,author])

df = pd.DataFrame(data, columns=["Quote", "Author"])

df.to_csv("quotes.csv", index=False)

print(df.head())
```

```python
all_quotes = []


for page in range(1,11):
  url = f"https://quotes.toscrape.com/page/{page}/"
  html = requests.get(url).text
  soup = BeautifulSoup(html, "html.parser")
  quotess = soup.find_all("div", class_="quote")


  for r in quotess:
    text = r.find ("span" , class_ = "text").text
    author = r.find ("small", class_ = "author").text
    all_quotes.append([text, author])

print(len(all_quotes))
```

```python
df = pd.DataFrame(all_quotes, columns=["Quote", "Author"])

df.to_csv("quotes.csv", index=False)

print(df.head(10))
```
