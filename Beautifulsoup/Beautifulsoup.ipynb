```python
import requests

base_url = "https://quotes.toscrape.com/page/"

# Loop through the first 3 pages
for i in range(1, 4):
  url = f"{base_url}{i}/"
  response = requests.get(url)

  # Print the HTML content for each page (optional, can be removed if not needed)
  print(f"--- Content for page {i} ---")
  print(response.text[:500]) # Print first 500 characters to avoid flooding output

  # Save the HTML content to a file
  with open (f"/content/sample_data/page{i}.html", "w") as f:
    f.write (response.text)
  print(f"Saved page {i} to /content/sample_data/page{i}.html")
```

```python
from bs4 import BeautifulSoup

all_authors = []

for i in range(1, 4): # Loop through pages 1, 2, 3
  file_path = f"/content/sample_data/page{i}.html"
  try:
    with open(file_path, "r") as f:
      content = f.read()

    soup = BeautifulSoup(content, "html.parser")
    author_elements = soup.find_all("small", class_="author")

    print(f"--- Authors from page {i} ---")
    for author_tag in author_elements:
      author_name = author_tag.get_text()
      print(author_name)
      all_authors.append(author_name)
  except FileNotFoundError:
    print(f"File not found: {file_path}")
  except Exception as e:
    print(f"An error occurred while processing {file_path}: {e}")

print("\n--- All unique authors found ---")
unique_authors = sorted(list(set(all_authors)))
for author in unique_authors:
  print(author)

print(f"\nTotal unique authors found: {len(unique_authors)}")
```

```python
all_tags = []

for i in range(1, 4): # Loop through pages 1, 2, 3
  file_path = f"/content/sample_data/page{i}.html"
  try:
    with open(file_path, "r") as f:
      content = f.read()

    soup = BeautifulSoup(content, "html.parser")
    # Find all 'div' elements with class 'tags'
    tags_div_elements = soup.find_all("div", class_="tags")

    print(f"--- Tags from page {i} ---")
    for tags_div in tags_div_elements:
      # Within each 'tags_div', find all 'a' elements with class 'tag'
      tag_elements = tags_div.find_all("a", class_="tag")
      for tag_element in tag_elements:
        tag_name = tag_element.get_text()
        print(tag_name)
        all_tags.append(tag_name)
  except FileNotFoundError:
    print(f"File not found: {file_path}")
  except Exception as e:
    print(f"An error occurred while processing {file_path}: {e}")

print("\n--- All unique tags found ---")
unique_tags = sorted(list(set(all_tags)))
for tag in unique_tags:
  print(tag)

print(f"\nTotal unique tags found: {len(unique_tags)}")
```

```python
all_quotes_data = []

for i in range(1, 4): # Loop through pages 1, 2, 3
  file_path = f"/content/sample_data/page{i}.html"
  try:
    with open(file_path, "r") as f:
      content = f.read()

    soup = BeautifulSoup(content, "html.parser")

    # Find all quote containers on the page
    quote_divs = soup.find_all("div", class_="quote")

    for quote_div in quote_divs:
      # Extract quote text
      quote_text = quote_div.find("span", class_="text").get_text()

      # Extract author
      author = quote_div.find("small", class_="author").get_text()

      # Extract tags (multiple tags per quote)
      tags_elements = quote_div.find("div", class_="tags").find_all("a", class_="tag")
      tags = [tag.get_text() for tag in tags_elements]

      all_quotes_data.append({"Quote": quote_text, "Author": author, "Tags": ", ".join(tags)})

  except FileNotFoundError:
    print(f"File not found: {file_path}")
  except Exception as e:
    print(f"An error occurred while processing {file_path}: {e}")

# Create a pandas DataFrame
df_quotes = pd.DataFrame(all_quotes_data)

# Save the DataFrame to an Excel file
excel_file_path = "quotes_data.xlsx"
df_quotes.to_excel(excel_file_path, index=False)

print(f"Successfully created Excel file: {excel_file_path}")
display(df_quotes.head())
```
