```python
import requests

base_url = "https://quotes.toscrape.com/page/"

# Loop through the first 3 pages
for i in range(4, 10):
  url = f"{base_url}{i}/"
  response = requests.get(url)

  # Print the HTML content for each page (optional, can be removed if not needed)
  print(f"--- Content for page {i} ---")
  print(response.text[:500]) # Print first 500 characters to avoid flooding output

  # Save the HTML content to a file
  with open (f"/content/sample_data/page{i}.html", "w") as f:
    f.write (response.text)
  print(f"Saved page {i} to /content/sample_data/page{i}.html")
```
